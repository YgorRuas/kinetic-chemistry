```python
"""
O Matrix Operator mexe com tudo oq é matriz. Tudo oq é parâmetro de bola aí é com a classe
"""


def matrix_operator(all_balls, balls, balls2, balls3):
    
    all_balls = np.concatenate((balls,balls2,balls3))
    all_balls_posx_list = [ball.posx for ball in all_balls]
    all_balls_posy_list = [ball.posy for ball in all_balls]
    all_balls_posx = np.array(all_balls_posx_list)
    all_balls_posy = np.array(all_balls_posy_list)
    all_balls_diameter_list = [ball.diameter for ball in all_balls]
    all_balls_diameter = np.array(all_balls_diameter_list)

    all_min_distances = (all_balls_diameter/2)[:, np.newaxis] + (all_balls_diameter/2)
    #garantindo que não checo duas vezes a mesma colisão
    all_min_distances[0:num_balls_white,num_balls_white:] = -np.inf
    
    # Calculate every combination using broadcasting
    all_dx = all_balls_posx[:, np.newaxis] - all_balls_posx
    all_dy = all_balls_posy[:, np.newaxis] - all_balls_posy
    
    all_distance_squared = all_dx**2 + all_dy**2
    all_distances = np.sqrt(all_distance_squared)

    #all_collision_index_x_list, all_collision_index_y_list =  np.where(np.less_equal(all_distances, all_min_distances))
    all_collisions_boolean_matrix = np.where((np.less_equal(all_distances, 0))==1, False, (np.less_equal(all_distances, all_min_distances)))
    all_reactions_boolean_matrix = all_collisions_boolean_matrix[balls.shape[0]:(balls.shape[0]+balls2.shape[0]),0:balls.shape[0]]

    
    # Suponha que você tenha a matriz all_collision_boolean_matrix
    n = all_collisions_boolean_matrix.shape[0]

    # Preencha a parte superior da matriz com False
    all_collisions_boolean_matrix[np.triu_indices(n, k=1)] = False
    all_collisions_index_x_list, all_collisions_index_y_list =  np.where(np.equal(all_collisions_boolean_matrix, True))
    
    all_reactions_index_x_list, all_reactions_index_y_list =  np.where(np.equal(all_collisions_boolean_matrix, True))
    
    

    collision_pairs = zip(all_collisions_index_x_list, all_collisions_index_y_list)
    reactions_pairs = zip(all_reactions_index_x_list, all_reactions_index_y_list)
    
    for pair in collision_pairs:
        all_balls[pair[0]].check_collision(all_balls[pair[1]])
        
    for pair in reactions_pairs:
            reacting_ball_0 = all_balls[pair[0]]
            reacting_ball_1 = all_balls[pair[1]]
            
            if reacting_ball_0.color == WHITE:
                if reacting_ball_1.color == YELLOW:
                    reacting_white_ball_index_plus = np.where(balls == reacting_ball_0)[0] #j
                    reacting_yellow_ball_index_plus = np.where(balls2 == reacting_ball_1)[0] #j1

                    reaction_value = random.uniform(0,100)
                    if reaction_value > 75 and reacting_white_ball_index_plus.size > 0 and reacting_yellow_ball_index_plus.size > 0:
                        reacting_white_ball_index = reacting_white_ball_index_plus[0]
                        reacting_yellow_ball_index = reacting_yellow_ball_index_plus[0]
                        balls2 = np.delete(balls2, reacting_yellow_ball_index, axis=0)
                        balls = np.delete(balls, reacting_white_ball_index, axis=0)
                        reacting_ball_0.check_reaction(reacting_ball_1)
                        balls3 = np.append(balls3, reacting_ball_0)


            if reacting_ball_0.color == YELLOW:
                if reacting_ball_1.color == WHITE:
                    reacting_white_ball_index_plus = np.where(balls == reacting_ball_1)[0] #j
                    reacting_yellow_ball_index_plus = np.where(balls2 == reacting_ball_0)[0] #j1

                    reaction_value = random.uniform(0,100)
                    if reaction_value > 75 and reacting_white_ball_index_plus.size > 0 and reacting_yellow_ball_index_plus.size > 0:
                        reacting_white_ball_index = reacting_white_ball_index_plus[0]
                        reacting_yellow_ball_index = reacting_yellow_ball_index_plus[0]
                        balls2 = np.delete(balls2, reacting_yellow_ball_index,axis=0)
                        balls = np.delete(balls, reacting_white_ball_index,axis=0)
                        reacting_ball_1.check_reaction(reacting_ball_0)
                        balls3 = np.append(balls3, reacting_ball_1)

  


    all_balls = np.concatenate((balls, balls2, balls3))

    return all_balls, balls, balls2, balls3
            
```

```python

pygame.init() #iniciando o pygame


# Parâmetros da tela de exibição 
WIDTH, SIMULATION_WIDTH, HEIGHT = 1280, 650, 650

# Itens do Menu
menu_items = ["Start Game", "Options", "Exit"]
selected_item = 0

fig, ax = plt.subplots()
xs = []
ys = []
y_limit = (0, 50)


MENU_FONT = pygame.font.Font(None, 36)
MENU_COLOR = (255, 255, 255)
MENU_SELECTED_COLOR = (255, 0, 0)
 

screen = pygame.display.set_mode((WIDTH, HEIGHT))

pygame.display.set_caption("Pong with Random Positions and Momentum Conservation")


 

def draw_menu():
    screen.fill((0, 0, 0))
    for i, item in enumerate(menu_items):
        color = MENU_SELECTED_COLOR if i == selected_item else MENU_COLOR
        text = MENU_FONT.render(item, True, color)
        text_rect = text.get_rect(center=(WIDTH // 2, HEIGHT // 2 + i * 50))
        screen.blit(text, text_rect)

        
def main_menu():

    global selected_item
    running = True

 

 

# Discretizando o tempo

clock = pygame.time.Clock() 
FPS = 100

 

# Cores padrões para exibição

BLACK =  (0,   0,   0  )
WHITE =  (255, 255, 255)
YELLOW = (255, 255, 0  )
BLUE =   (0,   0,   255)
RED =    (255, 0,   0  )

 

# Constantes físicas

k = 1.38e-23 # Constante de Boltzmann


#Classe Bolas Duras

class Ball:


    #Definindo o método para os parâmetros iniciais
    def __init__(self, posx, posy, diameter, speed, color, mass,temperature):
        self.posx = posx
        self.posy = posy
        self.diameter = diameter
        self.radius = diameter / 2
        self.color = color
        self.temperature = temperature
        self.mass = mass
        self.speed = speed
        self.xVel = self.speed * random.choice([-1, 1])
        self.yVel = self.speed * random.choice([-1, 1])
        self.velocity_measurements = []

        
    #Definindo o método de atualização
    def update(self):
        lx = None
        ly = None
        if self.check_wall_collision():
            lx = abs(self.xVel)*self.mass*2
            ly = abs(self.yVel)*self.mass*2
            #self.check_wall_collision()
        self.velocity_measurements.append(self.get_velocity())
        self.posx += self.xVel
        self.posy += self.yVel
        return lx, ly
    

        

    #Definindo o método para encontrar a velocidade
    def get_velocity(self):
        return math.sqrt(self.xVel**2 + self.yVel**2)


    #Definindo o método que checa colisões com a parede
    def check_wall_collision(self):
        
        if self.posy <= self.radius or self.posy >= HEIGHT - self.radius:
            self.yVel *= -1
            self.posy = max(self.radius, min(HEIGHT - self.radius, self.posy))  # Prevent sticking to walls
            return True

        if self.posx <= self.radius or self.posx >= SIMULATION_WIDTH - self.radius:
            self.xVel *= -1
            self.posx = max(self.radius, min(SIMULATION_WIDTH - self.radius, self.posx))  # Prevent sticking to walls
            return True

    def check_reaction(self, other_ball):

        self.color = BLUE
        self.mass = mass_blue
        self.diameter = diameter_blue
        

    #Definindo o método que checa colisões

    def check_collision(self, other_ball):

        dx = other_ball.posx - self.posx
        dy = other_ball.posy - self.posy
        distance_squared = dx**2 + dy**2
 

        if distance_squared <= (self.radius + other_ball.radius)**2:
            if distance_squared == 0:
                pass

            else:
                distance = math.sqrt(distance_squared)
                overlap = self.radius + other_ball.radius - distance
                overlap /= distance

                self.posx -= overlap * dx * 0.8
                self.posy -= overlap * dy * 0.8
                other_ball.posx += overlap * dx * 0.8
                other_ball.posy += overlap * dy * 0.8

            total_mass = self.mass + other_ball.mass
            self_momentum = self.mass * math.sqrt(self.xVel**2 + self.yVel**2)
            other_momentum = other_ball.mass * math.sqrt(other_ball.xVel**2 + other_ball.yVel**2)
            self_speed = self_momentum / self.mass
            other_speed = other_momentum / other_ball.mass
            relative_position = [other_ball.posx - self.posx, other_ball.posy - self.posy]
            relative_velocity = [other_ball.xVel - self.xVel, other_ball.yVel - self.yVel]
            relative_position_length = math.sqrt(relative_position[0]**2 + relative_position[1]**2)

            if relative_position_length != 0:
                dot_product = (relative_velocity[0] * relative_position[0] + relative_velocity[1] * relative_position[1]) / relative_position_length**2
                self.xVel += relative_position[0] * dot_product * 2 * other_ball.mass / total_mass
                self.yVel += relative_position[1] * dot_product * 2 * other_ball.mass / total_mass
                other_ball.xVel -= relative_position[0] * dot_product * 2 * self.mass / total_mass
                other_ball.yVel -= relative_position[1] * dot_product * 2 * self.mass / total_mass

   #Estrutura gráfica da bola dura         
    def display(self):
        pygame.draw.circle(screen, self.color, (int(self.posx), int(self.posy)), self.radius)


#Função principal    

def main(num_balls_white, num_balls_yellow, num_balls_blue, mass_white, mass_yellow, mass_blue, diameter_white, diameter_yellow, diameter_blue, temperature):
    running = True
    
    
    #balls = [Ball(random.randint(0 + diameter_white, SIMULATION_WIDTH - diameter_white), random.randint(0 + diameter_white, HEIGHT - diameter_white), diameter_white, speed_white, WHITE, mass_white, temperature) for _ in range(num_balls_white)]
    balls = np.array([Ball(random.randint(0 + diameter_white, SIMULATION_WIDTH - diameter_white), random.randint(0 + diameter_white, HEIGHT - diameter_white), diameter_white, speed_white, WHITE, mass_white, temperature) for _ in range(num_balls_white)])
    yellow_balls = np.array([Ball(random.randint(0 + diameter_yellow, SIMULATION_WIDTH - diameter_yellow), random.randint(0 + diameter_yellow, HEIGHT - diameter_yellow), diameter_yellow, speed_yellow, YELLOW, mass_yellow, temperature) for _ in range(num_balls_yellow)])
    blue_balls = np.array([Ball(random.randint(0 + diameter_blue, SIMULATION_WIDTH - diameter_blue), random.randint(0 + diameter_blue, HEIGHT - diameter_blue), diameter_blue, speed_blue, BLUE, mass_blue, temperature) for _ in range(num_balls_blue)])
   
    graph_update_counter = 0
    #white_graph = plot_velocity_distribution(balls, WHITE, "White Ball Velocity Distribution")
    #yellow_graph = plot_velocity_distribution(yellow_balls, YELLOW, "Yellow Ball Velocity Distribution")
    
    all_balls = np.concatenate((balls, yellow_balls,blue_balls))

    blue_concentration = []
    yellow_concentration = []
    white_concentration = []
    elapsed_time = []
    tempo = 0
    
    
    elo = []
    lxlx = []
    lyly = []
    pressurex = []
    pressurey = []

    # Aplicativo em funcionamento
    
    #while running:
    for i in range(0,1000):
        blue_concentration.append((blue_balls.size)/((12800*6500)))
        yellow_concentration.append((yellow_balls.size)/((12800*6500)))
        white_concentration.append((balls.size)/((12800*6500)))
        elapsed_time.append(tempo)

        screen.fill(BLACK) #Cor de fundo


        #Definindo modo para finalização
        for event in pygame.event.get(): 
            if event.type == pygame.QUIT:
                pygame.quit()
                running = False

                
        #Checando a colisão
        all_balls = np.concatenate((balls, yellow_balls,blue_balls))
        all_balls, balls, yellow_balls, blue_balls = matrix_operator(all_balls,balls,yellow_balls,blue_balls)
                
        
        # Individualizando as bolas
        for ball in all_balls:
            lx, ly = ball.update()
            ball.display()
            lxlx.append(lx)
            lyly.append(ly)
                

        pygame.display.update()
        clock.tick(FPS)
        tempo = tempo+(1/FPS)
        graph_update_counter += 1
    pygame.quit()
    return elapsed_time, white_concentration, yellow_concentration, blue_concentration, lxlx, lyly
 
    

if __name__ == "__main__":

 

    num_balls_white = 200  # Altera o número de bolas brancas
    num_balls_yellow = 200  # Altera o número de bolas amarelas
    num_balls_blue = 0
    mass_white = 3  # Altera o peso das bolas brancas
    mass_yellow = 1.5  # Altera o peso das bolas amarelas
    mass_blue = (mass_white/2)+mass_yellow
    diameter_white = 4  # Altera o diâmetro das bolas brancas
    diameter_yellow = 2  # Altera o diâmetro das bolas amarelas
    diameter_blue = diameter_white + diameter_yellow
    temperature = 320  # Altere isso para a temperatura desejada


    # Cálculo da velocidade inicial das bolas brancas com base na temperatura e peso
    speed_white = (math.sqrt(8 * k * temperature *10**20 / np.pi * mass_white))
    # Cálculo da velocidade inicial das bolas amarelas com base na temperatura e peso
    speed_yellow = (math.sqrt(8 * k * temperature *10**20 / np.pi * mass_yellow))
    # Cálculo da velocidade inicial das bolas azuis com base na temperatura e peso
    speed_blue = (math.sqrt(8 * k * temperature / np.pi * mass_blue))



    # Chamando a função principal para iniciar a simulação com os parâmetros especificados
    # Chamando a função principal para iniciar a simulação com os parâmetros especificados

    elapsed_time, white_concentration, yellow_concentration, blue_concentration, lxlx, lyly = main(num_balls_white, num_balls_yellow, num_balls_blue, mass_white, mass_yellow, mass_blue, diameter_white, diameter_yellow, diameter_blue, temperature)
```
